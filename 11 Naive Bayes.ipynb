# Naive Bayes - Titanic

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
target = df.Survived
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
X = df[["Pclass", "Sex" ,"Age", "Fare"]]
X
```

```python
pd.get_dummies(X.Sex)
```

```python
# dummies = pd.get_dummies(X.Sex, drop_first=True)
dummies = pd.get_dummies(X.Sex)
dummies
```

```python
X = pd.concat([X, dummies.female], axis=1)
X
```

```python
X.drop("Sex", axis=1, inplace=True)
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python
X.isna().any()
```

```python
X.columns[X.isna().any()]
```

```python
X.Age[:10]
```

## Imputación de los datos

```python
X.Age.mean()
```

```python
X.Age.median()
```

```python
X.Age.fillna(X.Age.median(), inplace=True)
```

```python
X.head(10)
```

```python
X.isnull().sum()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, target, test_size=0.25, random_state=23)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(X)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_train)
```

```python
model.score(X_test, y_test)
```

```python
np.mean(y_train == model.predict(X_train))
```

```python
np.mean(y_test == model.predict(X_test))
```
