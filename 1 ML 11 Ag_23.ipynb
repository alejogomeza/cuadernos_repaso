#EntropÃ­a

# **$$ S = -\sum_{n=1}^N  p_n\log_2  p_n $$**

**N** es el numero de estados.

**$p_n$** es la probabilidad de cada estado.
si se tiene dos estados las probabilidades son **$p$ y $(1-p)$**

#**$$ S = - pâ‹…g_2(p) - (1-p)â‹…\log_2(1-p) $$**

```python
# funcion de entropia
def S(p):
  return -p*np.log2(p) - (1-p)*np.log2(1-p)
```

```python
# aumjento la probabilidad de exito y bajo la entropia, para evaluar el modelo
S(0.627)
```

```python
#
S(0.876)
```

```python
# definir un epsilon para definir poco mas a 0
# ctrl + espacio para cargar la letra
Ïµ = 1e-3
p = np.linspace(0 + Ïµ, 1 - Ïµ)

plt.figure(figsize = (3, 3))
plt.plot(p, S(p))
plt.xlabel("Probabilidad")
plt.ylabel("EntropÃ­a")
plt.scatter(0.627, S(0.627), color = "red", label = "Primer\nexperimento")
plt.scatter(0.876, S(0.876), color = "green", label = "Segundo\nexperimento")
plt.legend(bbox_to_anchor = (1, 0 ,0.5 ,0.5))
plt.show()
```

## Proceso de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 23)
```

```python
x_train.shape , x_test.shape
```

```python
arbol2 = tree.DecisionTreeClassifier(criterion = "entropy",
                                     min_samples_split=50)
```

```python
arbol2.fit(x_train, y_train)
```

```python
plt.figure(figsize=(10, 6))
tree.plot_tree(arbol2,
               feature_names=["Perimetro", "Compacidad"],
               class_names=["B", "M"],
               filled=True)

plt.show()
```

Se evalua el entrenamiento

```python
y_pred_train = arbol2.predict(x_train)
```

```python
np.mean(y_train == y_pred_train)
```

Se evalua con los datos de prueba

```python
y_pred_test = arbol2.predict(x_test)
```

```python
np.mean(y_pred_test == y_test)
```

Matriz de confusiÃ³n - Train

```python
cm_train = confusion_matrix(y_train, y_pred_train)
cm_train
```

```python
fig, ax = plt.subplots(figsize=(5,5))

disp = ConfusionMatrixDisplay(confusion_matrix=cm_train,
                              display_labels = ["B", "M"])
disp.plot(ax = ax)

ax.set_title("Entrenamiento")
ax.set_xlabel("PredicciÃ³n")
ax.set_ylabel("Verdadero")

plt.show()
```

Matriz de confusiÃ³n - Test

```python
cm_test = confusion_matrix(y_test, y_pred_test)
cm_test
```

```python
fig, ax = plt.subplots(figsize=(5,5))

disp = ConfusionMatrixDisplay(confusion_matrix=cm_test,
                              display_labels = ["B", "M"])
disp.plot(ax = ax)

ax.set_title("Prueba")
ax.set_xlabel("PredicciÃ³n")
ax.set_ylabel("Verdadero")

plt.show()
```
