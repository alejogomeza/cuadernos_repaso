# Regresión Multilineal

Caso general

$$ y = f(x_1, x_2) = a x_1 + b x_2 + c $$

Caso particular

$$ y = f(x_1, x_2) = 2 x_1 + 3 x_2 + 4 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([ [1,1] , [1,2] , [2,3] , [2,2] ])
y = np.dot(X , np.array([2,3])) + 4
y
```

Construyo el modelo multinineal

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(np.array([[4,7]]))
```

# Support Vector Classification

```python
from sklearn import svm
```

```python
X = np.array([[0,0] , [1,1]])
y = np.array([0,1])
```

```python
clf = svm.SVC()
clf.fit(X,y)
```

```python
plt.figure(figsize=(3,3))
plt.scatter(X[np.where(y == 0)[0],0], X[np.where(y == 0)[0],1])
plt.scatter(X[np.where(y == 1)[0],0], X[np.where(y == 1)[0],1])
plt.show()
```

```python
p1 = [0.1 , 0.15]
p2 = [1.1 , 1.15]
clf.predict([p1, p2])
```

```python
p1 = [0.1 , 0.15]
p2 = [1.1 , 1.15]
color = ["navy", "salmon"]
plt.figure(figsize=(3,3))
plt.scatter(X[np.where(y == 0)[0],0], X[np.where(y == 0)[0],1], color = color[0])
plt.scatter(X[np.where(y == 1)[0],0], X[np.where(y == 1)[0],1], color = color[1])
plt.scatter(p1[0] , p1[1] , color = color[clf.predict([p1])[0]], marker="*")
plt.scatter(p2[0] , p2[1] , color = color[clf.predict([p2])[0]], marker="*")
plt.show()
```

```python
clf.support_vectors_
```

```python
clf.support_ # clasificación de los vectores de soporte
```

```python
clf.n_support_ # cuántos vectores de soporte hay de cada clase
```
